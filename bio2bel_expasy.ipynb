bio2bel_expasy library helps to download and pars the enzyme classes from the ExPASy ENZYME database.
It allows access to all important entries in the database via various query functions.
In addition it gives necessary functionality to enrich pybel.BELGraph graphs with relevant information.
To install it simply run 
# pip3 install git+https://gitlab.scai.fraunhofer.de/bio/PyExPASy.git
After that it can be included like this.

```python
from bio2bel_expasy.manager import Manager
```

After having this one can create a Manager() object which will parse all the data from expasy web servers and build database model out of it.

```python
m = Manager()
```

Call m.populate() to create the database. Allow some time for it.

```python
m.populate()
```

Now we can use standard getters for retrieving any information of our interest that in the database.
Here for the expasy id "child" we get the parent id.

```python
child = '1.1.1.1'
parent = m.get_parent(child)
print(parent)
```

For that parent id we can get all the child nodes by using get_children() method

```python
children = m.get_children(str(parent))
```

We can use ordinary foreach loop to process the data. Here we are printing ID's.

```python
for child_found in children:
    print(child_found)
```

With get_uniprot() method we can get UniProtKB/SwissProt ID's of proteins associated with the expasy entry.

```python
proteins = m.get_uniprot('1.1.1.1')
```

And can be processed in a foreach loop.

```python
for prot in proteins:
    print(prot)
```

Same for getting prosite ID's.

```python
prosites = m.get_prosite('1.1.1.1')
```

```python
for prosite_id in prosites:
    print(prosite_id)
```

Another very good functionality is the ability to enrich BELGraph objects with the data from expasy database.
For that first import pybel.BELGraph module and greate a BELGraph() object (see https://github.com/pybel/pybel for more details)

```python
from pybel import BELGraph
graph = BELGraph()
```

For proceeding further add the protein of your interest as a node into the graph.

```python
graph.add_simple_node('Protein', 'expasy', '1.1.1.1')
```

For enriching import enrich_enzyme_classes from bio2bel_expasy.enrich

```python
from bio2bel_expasy.enrich import enrich_enzyme_classes
```

Then simply call enrich_ynzyme_classes(graph).
Similar enrichment functions exist for enriching graph with the data from UniProtKB/SwissProt and Prosite, Parent, and child ID's (Note that enrich_enzyme_classes() does enrich for parent and child ID's).'

```python
graph = enrich_enzyme_classes(graph)
```

```python
for node in graph:
    print(node)
```
